#Task #05 作業 03【實作題】利用視覺化觀察資料

練習：請根據提供的 `iris` 資料，完成以下操作：

1. 利用資料視覺化觀察 sepal 和 petal 四個欄位的關係。
2. 利用相關係數觀察 sepal 和 petal 四個欄位的關係。

```python
# 請根據提供的 `iris` 資料，完成以下操作：
# 1. 利用資料視覺化觀察 sepal 和 petal 四個欄位的關係。
# 2. 利用相關係數觀察 sepal 和 petal 四個欄位的關係。

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
# 1. 利用資料視覺化觀察 sepal 和 petal 四個欄位的關係。
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

axes[0].plot(df['sepal length (cm)'], label='Sepal length')
axes[0].plot(df['sepal width (cm)'], label='Sepal width')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')
axes[0].legend()


axes[1].plot(df['petal length (cm)'], label='Petal length')
axes[1].plot(df['petal width (cm)'], label='Petal width')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Value')
axes[1].legend()

# 2. 利用相關係數觀察 sepal 和 petal 四個欄位的關係。

corr_of_iris = df[['sepal length (cm)', 'sepal width (cm)','petal length (cm)', 'petal width (cm)']].corr()

print(corr_of_iris)
```
